## **Lesson 1: The Mathematics of Phyllotaxis üåª**
## **Lecci√≥n 1: La Matem√°tica de la Filotaxis üåª**

In this project, you will learn to code the beautiful, intricate packing of florets in a sunflower head. This pattern, known as phyllotaxis, is not random‚Äîit is governed by a simple mathematical formula that uses the Golden Angle.

A formula can be used to describe the position of each floret:

$\theta = n \times \phi$  
$r = \sqrt{n}$

‚ÄãWhere:

* $n$ is the ordered number of each floret $(0,1,2,3...)$

* $\theta$ is the angle of the floret

* $\phi$ is the Golden Angle, approximately $137.5¬∞$

* $r$ is the floret's distance from the center of the sunflower

We can also use these values to find the $(x, y)$ coordinates of each floret using basic trigonometry. Remember that we will be working in **radians**, not degrees.

$x = r \times \cos(\theta)$  
$y = r \times \sin(\theta)$

En este proyecto, aprender√°s a programar el bello e intrincado empaquetamiento de las flores en un girasol. Este patr√≥n, conocido como filotaxis, no es aleatorio; est√° regido por una f√≥rmula matem√°tica simple que utiliza el √Ångulo Dorado.

Una f√≥rmula puede ser utilizada para describir la posici√≥n de cada fl√≥sculo:

Donde:

* $\theta$ es el √°ngulo del fl√≥sculo

* $\phi$ es el √Ångulo Dorado, aproximadamente $137.5¬∞$

* $r$ es la distancia del fl√≥sculo desde el centro del girasol

Tambi√©n podemos usar estos valores para encontrar las coordenadas $(x,y)$ de cada fl√≥sculo usando trigonometr√≠a b√°sica. Recuerda que trabajaremos en **radianes**, no en grados.

$x = r \times \cos(\theta)$  
$y = r \times \sin(\theta)$

---

### **Step 1: Set Up Our Code**
First, let's define the Golden Angle in radians, which is the value you'll use throughout this project. We'll also import the necessary libraries to perform our calculations and plot our results.

Run the cell below to set up your environment.

### **Paso 1: Preparar Nuestro C√≥digo**
Primero, vamos a definir el √Ångulo Dorado en radianes, que es el valor que usar√°s a lo largo de este proyecto. Tambi√©n importaremos las librer√≠as necesarias para realizar nuestros c√°lculos y graficar nuestros resultados.

Ejecuta la celda a continuaci√≥n para configurar tu entorno.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# The Golden Angle in degrees
golden_angle_deg = 137.50776405003785

# Convert the Golden Angle to radians
phi = golden_angle_deg * (math.pi / 180.0)

print(f"The Golden Angle in radians is: {phi:.4f}")
```

---
### **Step 2: Calculate Floret Coordinates**
Now, let's use the formulas to calculate the $(x, y)$ coordinates for a set of 1,000 florets. We'll store these coordinates in lists.

Run the code below to perform the calculations.

### **Paso 2: Calcular las Coordenadas del Fl√≥sculo**
Ahora, usemos las f√≥rmulas para calcular las coordenadas $(x,y)$ para un conjunto de 1,000 fl√≥sculos. Almacenaremos estas coordenadas en listas.

Ejecuta el c√≥digo a continuaci√≥n para realizar los c√°lculos.

```python
num_florets = 1000

# Create empty lists to store the coordinates
x_coords = []
y_coords = []

# Loop through each floret number (n)
for n in range(num_florets):
    # Calculate the radius and angle for this floret
    r = math.sqrt(n)
    theta = n * phi
    
    # Calculate the x and y coordinates and append them to the lists
    x = r * math.cos(theta)
    y = r * math.sin(theta)
    
    x_coords.append(x)
    y_coords.append(y)
```

## **Lesson 2: The Growing Sunflower üåª**
## **Lecci√≥n 2: El Girasol en Crecimiento üåª**

While a static image is a great start, a sunflower is a living, growing plant. The florets arise from the center of the sunflower and are pushed outward as new ones are formed. In this lesson, we'll create an animation that simulates this growth.

The key to this animation is a two-part process:

1. We'll pre-calculate the positions for all the florets we plan to grow.

2. In a loop, we'll add one new floret at a time. The new floret will have the shortest radius, and all the previous florets will be "pushed" outward, getting a slightly longer radius.

To make this work, we'll create lists of angles and radii for all our florets, then use a for loop to control the animation. In each iteration of the loop, a new floret is added, and the radii of all existing florets are adjusted to simulate growth.

Aunque una imagen est√°tica es un gran comienzo, un girasol es una planta viva y en crecimiento. Los fl√≥sculos surgen del centro del girasol y son empujados hacia afuera a medida que se forman otros nuevos. En esta lecci√≥n, crearemos una animaci√≥n que simula este crecimiento.

La clave para esta animaci√≥n es un proceso de dos partes:

1. Calcularemos previamente las posiciones para todos los fl√≥sculos que planeamos hacer crecer.

2. En un bucle, agregaremos un nuevo fl√≥sculo a la vez. El nuevo fl√≥sculo tendr√° el radio m√°s corto, y todos los fl√≥sculos anteriores ser√°n "empujados" hacia afuera, obteniendo un radio ligeramente m√°s largo.

Para que esto funcione, crearemos listas de √°ngulos y radios para todos nuestros fl√≥sculos, luego usaremos un bucle for para controlar la animaci√≥n. En cada iteraci√≥n del bucle, se agrega un nuevo fl√≥sculo y los radios de todos los fl√≥sculos existentes se ajustan para simular el crecimiento.


---
### **Step 1: Create Pre-Populated Lists**
First, let's create two lists, one for thetas (angles) and one for radii, containing the values for all the florets we want to grow in our animation. We will calculate the values for 750 florets.

### **Paso 1: Crear Listas Pre-Pobladas**
Primero, vamos a crear dos listas, una para thetas (√°ngulos) y otra para radios, que contengan los valores para todos los fl√≥sculos que queremos que crezcan en nuestra animaci√≥n. Calcularemos los valores para 750 fl√≥sculos.

```python
# The golden angle, as before
phi = 137.50776405003785 * (math.pi / 180.0)

# Create pre-populated lists for theta and radii
thetas = []
radii = []

num_florets_to_grow = 750

for n in range(num_florets_to_grow):
    r = math.sqrt(n)
    theta = n * phi
    
    thetas.append(theta)
    radii.append(r)
```

---
### **Step 2: Animate the Growth**
Now for the fun part! We'll use a for loop to simulate the growth of the sunflower. With each iteration, we'll calculate the x and y coordinates for the current frame, ensuring that the oldest florets have the longest radii.

### **Paso 2: Animar el Crecimiento**
Ahora viene la parte divertida. Usaremos un bucle for para simular el crecimiento del girasol. Con cada iteraci√≥n, calcularemos las coordenadas x e y para el cuadro actual, asegur√°ndonos de que los fl√≥sculos m√°s antiguos tengan los radios m√°s largos.

```python
# --- Step 2: Animate the Growth ---
import time
import matplotlib.pyplot as plt
import math
from IPython.display import display, clear_output

# Create a figure for the animation
fig = plt.figure(figsize=(5, 5))

for i in range(len(radii)):
    x_coords = []
    y_coords = []
    
    # This selects the current set of thetas to use
    current_thetas = thetas[0:(i + 1)]
    # This selects the current set of radii to use
    current_radii = radii[0:(i + 1)]
    
    # After creating the list of radii, "reverse" its order.
    # This ensures that the florets added first have the longest radius.
    current_radii.reverse() # This will now work because radii is a Python list

    for n in range(i + 1):
        # Here, for each floret n, we calculate r, theta, x, and y.
        r = current_radii[n]
        theta = current_thetas[n]

        x = r * math.cos(theta)
        y = r * math.sin(theta)

        x_coords.append(x)
        y_coords.append(y)

    # Put your code to make a plot of a sunflower here
    plt.scatter(x_coords, y_coords, alpha=0.3, c="gray", s=50)
    plt.scatter(x_coords[::5], y_coords[::5], c="orange", s=50)
    plt.scatter(x_coords[::8], y_coords[::8], c="magenta", s=50)
    plt.scatter(x_coords[::13], y_coords[::13], c="yellow", s=50)
    plt.scatter(x_coords[::21], y_coords[::21], c="deepskyblue", s=50)
    plt.scatter(x_coords[::34], y_coords[::34], c="green", s=50)

    plt.xlim(-30, 30)
    plt.ylim(-30, 30)
    plt.axis('off')
    plt.axis('equal')

    # This is the code that creates the animation
    time.sleep(0.001)
    clear_output(wait=True)
    display(fig)
    fig.clear()

# Closes the figure animation once complete
plt.close()
```

## **Lesson 3: Transforming the Leaves üåø**
## **Lecci√≥n 3: Transformando las Hojas üåø**

Our leaf outlines, as beautiful as they are, aren't ready for our animation just yet. The data files were created with the leaves pointing downward, and their size and position aren't standardized. To use them in our spiral, we need to apply a series of transformations.

Nuestros contornos de hojas, por muy bonitos que sean, a√∫n no est√°n listos para nuestra animaci√≥n. Los archivos de datos se crearon con las hojas apuntando hacia abajo, y su tama√±o y posici√≥n no est√°n estandarizados. Para usarlas en nuestra espiral, necesitamos aplicar una serie de transformaciones.

---
### **Step 1: The Leaf Data Files**
All our leaf outlines are stored in the leaf_shapes directory as simple, tab-delimited text files. Each file contains two columns: one for x-coordinates and one for y-coordinates.

We've used a specific naming convention:

* The _main.txt files (e.g., agave_main.txt) contain the data for the primary part of the leaf.

* The _accent.txt files (e.g., cempasuchil_accent.txt) contain data for any secondary, inner shape.

This naming system allows our code to automatically find the right files for each plant. Note that you only need a _main.txt file, the _accent.txt file is optional.

### **Paso 1: Los Archivos de Datos de las Hojas**
Todos nuestros contornos de hojas se almacenan en el directorio leaf_shapes como archivos de texto sencillos y delimitados por tabulaciones. Cada archivo contiene dos columnas: una para las coordenadas x y otra para las coordenadas y.

Hemos utilizado una convenci√≥n de nombres espec√≠fica:

* Los archivos _main.txt (por ejemplo, agave_main.txt) contienen los datos de la parte principal de la hoja.

* Los archivos _accent.txt (por ejemplo, cempasuchil_accent.txt) contienen datos para cualquier forma secundaria o interior.

Este sistema de nombres permite que nuestro c√≥digo encuentre autom√°ticamente los archivos correctos para cada planta. Ten en cuenta que solo necesitas un archivo _main.txt; el archivo _accent.txt es opcional.

```python
# --- Main Transformation Logic ---
leaf_data_list = [
    {"prefix": "cempasuchil_main", "accent_prefix": "cempasuchil_accent", "colors": ["#FF9C00", "#FF5B00"]},
    {"prefix": "agave_main", "accent_prefix": None, "colors": ["#6D9DB3"]},
    {"prefix": "aloe_main", "accent_prefix": None, "colors": ["#68989E"]},
    {"prefix": "succulent_main", "accent_prefix": "succulent_accent", "colors": ["#38a28b", "#fd3b5d"]},
]

def load_2d_leaf_data(name_prefix, output_folder="leaf_shapes"):
    """
    Loads 2D coordinate data from a text file.
    Returns x and y coordinates as NumPy arrays.
    """
    filepath = os.path.join(output_folder, f"{name_prefix}.txt")
    try:
        data_array = np.loadtxt(filepath, dtype=float, delimiter='\t')
        x_coords = data_array[:, 0]
        y_coords = data_array[:, 1]
        return x_coords, y_coords
    except (FileNotFoundError, IndexError, ValueError):
        return None, None
```

---
### **Step 2: The transform_leaf_data Function**
This function is the heart of our data preparation. It takes our raw leaf coordinates and performs three essential transformations: flipping, translating, and scaling.

### **Paso 2: La Funci√≥n transform_leaf_data**
Esta funci√≥n es el coraz√≥n de nuestra preparaci√≥n de datos. Toma las coordenadas crudas de nuestras hojas y realiza tres transformaciones esenciales: volteo, traslaci√≥n y escalado.

```python
def transform_leaf_data(main_x, main_y, accent_x=None, accent_y=None):
    """
    Flips, translates, and scales leaf data to a standard upright position
    at the origin (0,0) with a max length of 1.
    """
    if main_x is None or main_y is None:
        return None, None, None, None

    # Step 1: Flip the leaf upright. Our original data is upside-down,
    # so we negate the y-coordinates to flip it correctly.
    reflected_main_x = np.copy(main_x)
    reflected_main_y = -np.copy(main_y)
    
    # Do the same for the accent shape, if it exists.
    reflected_accent_x, reflected_accent_y = None, None
    if accent_x is not None and accent_y is not None:
        reflected_accent_x = np.copy(accent_x)
        reflected_accent_y = -np.copy(accent_y)

    # Step 2: Translate the leaf to the origin (0,0).
    # We find the bottom-most point of the leaf and shift it to the origin.
    min_y = np.min(reflected_main_y)
    base_x_coords = reflected_main_x[np.where(reflected_main_y == min_y)]
    base_x = np.mean(base_x_coords)
    
    transformed_main_x = reflected_main_x - base_x
    transformed_main_y = reflected_main_y - min_y
    
    transformed_accent_x, transformed_accent_y = None, None
    if reflected_accent_x is not None and reflected_accent_y is not None:
        transformed_accent_x = reflected_accent_x - base_x
        transformed_accent_y = reflected_accent_y - min_y
        
    # Step 3: Scale the leaf to a standard length of 1.
    # This makes it easy to control the size of each leaf in the animation.
    if transformed_main_y is not None and len(transformed_main_y) > 0:
        leaf_length = np.max(transformed_main_y)
        if leaf_length > 0:
            transformed_main_x /= leaf_length
            transformed_main_y /= leaf_length
            if transformed_accent_x is not None and transformed_accent_y is not None:
                transformed_accent_x /= leaf_length
                transformed_accent_y /= leaf_length

    return transformed_main_x, transformed_main_y, transformed_accent_x, transformed_accent_y
```

---
### **Step 3: Building the Master Dictionary**
This loop acts as a processing "factory," applying our functions to every plant in the leaf_data_list and storing the results in a new dictionary called transformed_data.

### **Paso 3: Construir el Diccionario Maestro**
Este bucle act√∫a como una "f√°brica" de procesamiento, aplicando nuestras funciones a cada planta en la lista leaf_data_list y almacenando los resultados en un nuevo diccionario llamado transformed_data.

```python
import os

transformed_data = {}
for leaf in leaf_data_list:
    # Load the raw data for the main and accent shapes
    main_x, main_y = load_2d_leaf_data(leaf["prefix"])
    accent_x, accent_y = None, None
    if leaf["accent_prefix"]:
        accent_x, accent_y = load_2d_leaf_data(leaf["accent_prefix"])
        
    # Transform the data using our function
    trans_main_x, trans_main_y, trans_accent_x, trans_accent_y = transform_leaf_data(main_x, main_y, accent_x, accent_y)
    
    # Store the transformed data in our dictionary
    transformed_data[leaf["prefix"].replace("_main", "").title()] = {
        "main_x": trans_main_x, "main_y": trans_main_y,
        "accent_x": trans_accent_x, "accent_y": trans_accent_y,
        "colors": leaf["colors"]
    }
```

Note that in the previous coding cell, the for loop automatically extracts the type of plant from the names of the files themselves so that you can choose which plant to analyze, like "Cempasuchil", "Agave", "Succulent", "Aloe", etc. (note that the first letter is capitalized).

###  **Paso 4: Visualizaci√≥n: Antes y Despu√©s**

## **Lesson 4: Animating the Growing Plant ü™¥**
## **Lecci√≥n 4: Animando la Planta en Crecimiento ü™¥**
In this lesson, we will combine our transformed leaf data with the phylotaxis algorithm to animate a growing plant. We'll use a for loop to draw each new leaf, creating a sense of natural growth.

En esta lecci√≥n, combinaremos nuestros datos de hojas transformadas con el algoritmo de filotaxis para animar una planta en crecimiento. Utilizaremos un bucle for para dibujar cada nueva hoja, creando una sensaci√≥n de crecimiento natural.

---
### **Step 1: The Adjustable Parameters**
At the top of the code, we'll set up a section for adjustable parameters. This makes it easy for you to control the animation's look and feel without changing the core logic. Each parameter is clearly named and commented to explain its function.

NUM_LEAVES: Determines the final number of leaves in the animation. More leaves mean a denser, more complete plant.

FRAME_DURATION_MS: Controls how long each frame of the GIF is displayed, in milliseconds. A smaller number creates a faster animation.

LEAF_OUTLINE_WIDTH: Sets the thickness of the black outline on the main leaves, which helps them stand out.

The FADE and GROWTH parameters control how the leaves change as they grow outward from the center, creating a realistic effect.

### **Paso 1: Los Par√°metros Ajustables**
En la parte superior del c√≥digo, configuraremos una secci√≥n para los par√°metros ajustables. Esto facilita el control de la apariencia de la animaci√≥n sin cambiar la l√≥gica principal. Cada par√°metro tiene un nombre claro y est√° comentado para explicar su funci√≥n.

NUM_LEAVES: Determina el n√∫mero final de hojas en la animaci√≥n. M√°s hojas significan una planta m√°s densa y completa.

FRAME_DURATION_MS: Controla cu√°nto tiempo se muestra cada cuadro del GIF, en milisegundos. Un n√∫mero m√°s peque√±o crea una animaci√≥n m√°s r√°pida.

LEAF_OUTLINE_WIDTH: Establece el grosor del contorno negro en las hojas principales, lo que ayuda a que se destaquen.

Los par√°metros de FADE (atenuaci√≥n) y GROWTH (crecimiento) controlan c√≥mo cambian las hojas a medida que crecen desde el centro, creando un efecto realista.

---
### **Step 2: The Animation Loop**
This is the core of the animation. The main for loop iterates from 0 to NUM_LEAVES, adding one leaf to the plot in each iteration.

ax.clear(): This command is crucial. It clears the plot at the beginning of each frame so we can draw the new set of leaves from scratch.

Phyllotaxis: Inside the loop, we use a second nested for loop to draw each leaf. We apply the golden angle and square root of n to determine the radial position (r) and angle (theta) of each leaf, creating the spiral pattern.

Transformations: We then use math.cos and math.sin to rotate each leaf to its correct position on the spiral. We also apply dynamic scaling and fading based on its radius to simulate the leaves growing and aging.

ax.fill(): This command is what draws the leaf shapes. It takes the rotated coordinates and colors them in. We also use edgecolor and linewidth to add the black outline you requested.

### **Paso 2: El Bucle de Animaci√≥n**
Este es el n√∫cleo de la animaci√≥n. El bucle for principal itera desde 0 hasta NUM_LEAVES, agregando una hoja al gr√°fico en cada iteraci√≥n.

ax.clear(): Este comando es crucial. Limpia el gr√°fico al comienzo de cada cuadro para que podamos dibujar el nuevo conjunto de hojas desde cero.

Filotaxis: Dentro del bucle, utilizamos un segundo bucle for anidado para dibujar cada hoja. Aplicamos el √°ngulo dorado y la ra√≠z cuadrada de n para determinar la posici√≥n radial (r) y el √°ngulo (theta) de cada hoja, creando el patr√≥n en espiral.

Transformaciones: Luego usamos math.cos y math.sin para rotar cada hoja a su posici√≥n correcta en la espiral. Tambi√©n aplicamos escalado din√°mico y atenuaci√≥n basados en su radio para simular el crecimiento y envejecimiento de las hojas.

ax.fill(): Este comando es el que dibuja las formas de las hojas. Toma las coordenadas rotadas y las colorea. Tambi√©n usamos edgecolor y linewidth para agregar el contorno negro que solicitaste.

---
### **Step 3: Saving the GIF**
After the animation loop, we use the Pillow library to save all the collected frames as a single, looping GIF. The save_all and append_images parameters ensure that every frame is included in the final animation.

Here is the complete code.

### **Paso 3: Guardar el GIF**
Despu√©s del bucle de animaci√≥n, usamos la librer√≠a Pillow para guardar todos los cuadros recopilados como un √∫nico GIF en bucle. Los par√°metros save_all y append_images aseguran que cada cuadro se incluya en la animaci√≥n final.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import math
import time
import io

from PIL import Image
from IPython.display import display, clear_output

# --- ADJUSTABLE PARAMETERS ---
NUM_LEAVES = 100
IMAGE_DPI = 72
FRAME_DURATION_MS = 150
FADE_START_RADIUS_PROPORTION = 0.4  # Petals start to fade at 40% of the max radius
FADE_END_RADIUS_PROPORTION = 0.5    # Petals are fully faded at 50% of the max radius
GROWTH_PLATEAU_PROPORTION = 0.5
LEAF_OUTLINE_WIDTH = 0.0            # Thin black outline for main leaves. Select 0 for no outline
leaf_shape_name = "Cempasuchil"      # Change this to switch plants (e.g., "Cempasuchil", "Agave", "Aloe", "Succulent")

# --- CONSTANTS (Shouldn't need to change) ---
golden_angle = 137.5077640500378546463487396283702776206886952699253696312384958261062333851951
phi = golden_angle * (math.pi / 180.0)
leaf_data_dir = "leaf_shapes"

# --- Main Transformation Logic ---
leaf_data_list = [
    {"prefix": "cempasuchil_main", "accent_prefix": "cempasuchil_accent", "colors": ["#FF9C00", "#FF5B00"]},
    {"prefix": "agave_main", "accent_prefix": None, "colors": ["#6D9DB3"]},
    {"prefix": "aloe_main", "accent_prefix": None, "colors": ["#68989E"]},
    {"prefix": "succulent_main", "accent_prefix": "succulent_accent", "colors": ["#38a28b", "#fd3b5d"]},
]

def load_2d_leaf_data(name_prefix, output_folder=leaf_data_dir):
    """Loads 2D coordinate data from a text file."""
    filepath = os.path.join(output_folder, f"{name_prefix}.txt")
    try:
        data_array = np.loadtxt(filepath, dtype=float, delimiter='\t')
        x_coords = data_array[:, 0]
        y_coords = data_array[:, 1]
        return x_coords, y_coords
    except (FileNotFoundError, IndexError, ValueError):
        return None, None

def transform_leaf_data(main_x, main_y, accent_x=None, accent_y=None):
    """Flips, translates, and scales leaf data."""
    if main_x is None or main_y is None:
        return None, None, None, None
    reflected_main_x = np.copy(main_x)
    reflected_main_y = -np.copy(main_y)
    reflected_accent_x, reflected_accent_y = None, None
    if accent_x is not None and accent_y is not None:
        reflected_accent_x = np.copy(accent_x)
        reflected_accent_y = -np.copy(accent_y)
    min_y = np.min(reflected_main_y)
    base_x_coords = reflected_main_x[np.where(reflected_main_y == min_y)]
    base_x = np.mean(base_x_coords)
    transformed_main_x = reflected_main_x - base_x
    transformed_main_y = reflected_main_y - min_y
    transformed_accent_x, transformed_accent_y = None, None
    if reflected_accent_x is not None and reflected_accent_y is not None:
        transformed_accent_x = reflected_accent_x - base_x
        transformed_accent_y = reflected_accent_y - min_y
    if transformed_main_y is not None and len(transformed_main_y) > 0:
        leaf_length = np.max(transformed_main_y)
        if leaf_length > 0:
            transformed_main_x /= leaf_length
            transformed_main_y /= leaf_length
            if transformed_accent_x is not None and transformed_accent_y is not None:
                transformed_accent_x /= leaf_length
                transformed_accent_y /= leaf_length
    return transformed_main_x, transformed_main_y, transformed_accent_x, transformed_accent_y

transformed_data = {}
for leaf in leaf_data_list:
    main_x, main_y = load_2d_leaf_data(leaf["prefix"])
    accent_x, accent_y = None, None
    if leaf["accent_prefix"]:
        accent_x, accent_y = load_2d_leaf_data(leaf["accent_prefix"])
    trans_main_x, trans_main_y, trans_accent_x, trans_accent_y = transform_leaf_data(main_x, main_y, accent_x, accent_y)
    transformed_data[leaf["prefix"].replace("_main", "").title()] = {
        "main_x": trans_main_x, "main_y": trans_main_y,
        "accent_x": trans_accent_x, "accent_y": trans_accent_y,
        "colors": leaf["colors"]
    }

# --- PHYLOTAXIS & ANIMATION LOGIC ---
thetas = np.zeros(NUM_LEAVES)
radii = np.zeros(NUM_LEAVES)
for n in range(NUM_LEAVES):
    radii[n] = math.sqrt(n)
    thetas[n] = n * phi

leaf_shape = transformed_data.get(leaf_shape_name)
if leaf_shape is None:
    print(f"Error: Leaf shape '{leaf_shape_name}' not found.")
    exit()

trans_main_x, trans_main_y = leaf_shape["main_x"], leaf_shape["main_y"]
trans_accent_x, trans_accent_y = leaf_shape["accent_x"], leaf_shape["accent_y"]
main_color = leaf_shape["colors"][0]
accent_color = leaf_shape["colors"][1] if len(leaf_shape["colors"]) > 1 else None

# Calculate dynamic parameters based on NUM_LEAVES
MAX_RADIUS = radii[-1]
FADE_START_RADIUS = MAX_RADIUS * FADE_START_RADIUS_PROPORTION
FADE_END_RADIUS = MAX_RADIUS * FADE_END_RADIUS_PROPORTION
GROWTH_PLATEAU_START = MAX_RADIUS * GROWTH_PLATEAU_PROPORTION
GROWTH_MAX_SIZE = MAX_RADIUS * 0.9

# --- Main Animation Loop ---
fig, ax = plt.subplots(figsize=(6, 6), dpi=IMAGE_DPI)
ax.set_facecolor("black")
fig.patch.set_facecolor("black")
ax.set_aspect('equal', 'box')
ax.axis('off')
ax.set_xlim(-MAX_RADIUS - 1, MAX_RADIUS + 1)
ax.set_ylim(-MAX_RADIUS - 1, MAX_RADIUS + 1)

frames = []

for i in range(NUM_LEAVES):
    ax.clear()
    ax.set_facecolor("black")
    ax.set_aspect('equal', 'box')
    ax.axis('off')
    ax.set_xlim(-MAX_RADIUS - 1, MAX_RADIUS + 1)
    ax.set_ylim(-MAX_RADIUS - 1, MAX_RADIUS + 1)

    current_thetas = thetas[:i+1]
    current_radii = radii[:i+1]
    current_radii_reversed = current_radii[::-1]

    for j in range(len(current_radii_reversed)):
        r = current_radii_reversed[j]
        theta = current_thetas[j]
        
        # Growth and fading logic
        if r <= GROWTH_PLATEAU_START:
            leaf_r = np.interp(r, [0, GROWTH_PLATEAU_START], [0, GROWTH_MAX_SIZE])
        else:
            leaf_r = np.interp(r, [GROWTH_PLATEAU_START, FADE_START_RADIUS], [GROWTH_MAX_SIZE, GROWTH_MAX_SIZE])

        if r <= FADE_START_RADIUS:
            alpha = 1
        elif r > FADE_END_RADIUS:
            alpha = 0
        else:
            alpha = np.interp(r, [FADE_START_RADIUS, FADE_END_RADIUS], [1.0, 0.0])
            
        petal_angle = theta
        cos_petal = np.cos(petal_angle)
        sin_petal = np.sin(petal_angle)

        # --- Moved rotation calculations here for correct scope ---
        scaled_main_x = trans_main_x * leaf_r
        scaled_main_y = trans_main_y * leaf_r
        rotated_main_x = scaled_main_x * cos_petal - scaled_main_y * sin_petal
        rotated_main_y = scaled_main_y * cos_petal + scaled_main_x * sin_petal

        if trans_main_x is not None and trans_main_y is not None:
            ax.fill(rotated_main_x, rotated_main_y, color=main_color, alpha=alpha, zorder=j,
                    edgecolor='black', linewidth=LEAF_OUTLINE_WIDTH)
        
        if trans_accent_x is not None and trans_accent_y is not None:
            scaled_accent_x = trans_accent_x * leaf_r
            scaled_accent_y = trans_accent_y * leaf_r
            rotated_accent_x = scaled_accent_x * cos_petal - scaled_accent_y * sin_petal
            rotated_accent_y = scaled_accent_y * cos_petal + scaled_accent_x * sin_petal
            ax.fill(rotated_accent_x, rotated_accent_y, color=accent_color, alpha=alpha, zorder=j)

    buf = io.BytesIO()
    plt.savefig(buf, format='png', dpi=IMAGE_DPI, bbox_inches='tight', pad_inches=0)
    buf.seek(0)
    image = Image.open(buf)
    frames.append(image)
    
    display(fig)
    clear_output(wait=True)
    
plt.close(fig)

if frames:
    frames[0].save('animation.gif',
                   save_all=True,
                   append_images=frames[1:],
                   optimize=True,
                   duration=FRAME_DURATION_MS,
                   loop=0)
    print("Animation saved as animation.gif")
```
